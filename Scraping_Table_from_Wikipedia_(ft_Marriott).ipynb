```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

wikiurl="https://en.wikipedia.org/wiki/Marriott_International"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
table=soup.find('table',{'class':"wikitable"})
```

```python
table
```

```python
df=pd.read_html(str(table))
# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head())
```

```python
df
```

```python
df = df.rename(columns={'Total Assetsin mil. USD$': 'Total Assets (in US$ mil)', 'Revenuein mil. USD$': 'Revenue (in US$ mil)', 'Net incomein mil. USD$':"Net income (in US$ mil)"}) 
```

```python
fig = make_subplots(rows=1, cols=4, shared_xaxes=True, shared_yaxes=False, start_cell='top-left', x_title="Year", y_title="US Dollar")


revenue = df1.iloc[:, 0]
netincome = df1.iloc[:, 1]
asset = df1.iloc[:, 2]
employees = df1.iloc[:, 3]

fig.add_trace(go.Scatter(x=df1.index, y=revenue, name='Revenue'), row=1, col=1)
fig.add_trace(go.Scatter(x=df1.index, y=netincome, name='Net Income'), row=1, col=2)
fig.add_trace(go.Scatter(x=df1.index, y=asset, name='Total Assets'), row=1, col=3)
fig.add_trace(go.Scatter(x=df1.index, y=employees, name='Employees'), row=1, col=4)
fig.show()
```
